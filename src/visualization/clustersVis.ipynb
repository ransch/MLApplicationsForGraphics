```
import pickle
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from src import settings
```

```
clusters = [0,1,3,5,10,55,57,63,95,101,105]
numEach = 5
```

```
frogsInds = settings.frogs6000
with open(settings.p / 'clustering/6000-dim-100-clst-128/clusters.pkl', 'rb') as f:
  pcklr = pickle.Unpickler(f)
  buckets, _ = pcklr.load()
```

```
fig, axs = plt.subplots(len(clusters), numEach, figsize=(15,15))
for cluster_ind in range(len(clusters)):
  cluster = clusters[cluster_ind]
  for i in range(min(numEach, len(buckets[cluster]))):
    ind = buckets[cluster][i]
    img = mpimg.imread(str(settings.frogs / f'frog-{frogsInds[ind]}.png'))
    axs[cluster_ind, i].imshow(img)
```
